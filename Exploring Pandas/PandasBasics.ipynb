```python
import pandas as pd
import numpy as np
```

```python
data ={'planet':['Mercury','Venus','Earth','Mars','Jupiter','Saturn','Uranus','Neptune'],
       'moons':['0','0','1','2','80','83','27','14'],
       'radius_km':['2400','6052','6371','3390','69911','582323','25362','24622']
       }
planets = pd.DataFrame(data)
planets          
```

```python
#Note information was loaded as String type
# mask = planets['moons'] < 20 -> will trough TypeError
mask = planets['moons'] < '20'
mask 
```

```python
#convert string into int
planets['moons']=planets['moons'].astype(int)
planets['radius_km']=planets['radius_km'].astype(int)
planets
```

```python
# Create a Boolean mask of planets with more than 20 moons, excluding them if they
# have 80 moons or if their radius is less than 50,000 km.
mask = (planets['moons'] > 20 ) & (planets['moons'] <81) & (planets['radius_km'] < 50000) 
planets[mask]
```

```python
#Basic consultation to a DateFrame
planets.describe()
planets.head(10)
planets.shape
planets.info
planets.ndim
planets.columns
planets.groupby(['moons']).sum()
```

```python
# Instantiate a dictionary of planetary data.
data2 = {'type': ['terrestrial', 'terrestrial', 'terrestrial', 'terrestrial',
                 'gas giant', 'gas giant', 'ice giant', 'ice giant'],
        'rings': ['no', 'no', 'no', 'no', 'yes', 'yes', 'yes','yes'],
        'mean_temp_c': [167, 464, 15, -65, -110, -140, -195, -200],
        'magnetic_field': ['yes', 'no', 'yes', 'no', 'yes', 'yes', 'yes', 'yes']
        }

# Use pd.DataFrame() function to convert dictionary to dataframe.
planets2 = pd.DataFrame(data2)
planets2
```

```python
#concatenate
planets_df=pd.concat([planets,planets2], axis=0)
planets_df
```

```python
# Reset the row indices.
df3 = planets_df.reset_index(drop=True)
df3
```

```python
clothes = pd.DataFrame({'type': ['pants', 'shirt', 'shirt', 'pants', 'shirt', 'pants'],
                       'color': ['red', 'blue', 'green', 'blue', 'green', 'red'],
                       'price_usd': [20, 35, 50, 40, 100, 75],
                       'mass_g': [125, 440, 680, 200, 395, 485]})
clothes
```

```python
grouped = clothes.groupby('type')
grouped.mean(numeric_only=True)
clothes.groupby(['type', 'color']).size()
grouped.sum(numeric_only=True)
grouped.std(numeric_only=True)
grouped.count() #not nulls
grouped.min()
```

```python
#function agg
clothes[['price_usd', 'mass_g']].agg(['sum', 'mean'], axis=0)

clothes.groupby('color').agg({'price_usd': ['mean', 'max'],
                             'mass_g': ['mean', 'max']})
```
